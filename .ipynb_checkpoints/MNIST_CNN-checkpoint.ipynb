### Importações de bibliotecas

```python
#BIBLIOTECAS PARA MANIPULAÇÃO E VISUALIZAÇÃO DE DADOS
import numpy as np
import matplotlib.pyplot as plt

#BIBLIOTECAS PYTORCH
import torch
import torch.nn as nn
from torch import optim
import torch.nn.functional as F
from torchvision import datasets
import torchvision.transforms as transforms
print(torch.__version__)
```

### Base de dados

```python
#Setamos o seed para obtermos os mesmos resultados
torch.manual_seed(123)
```

```python
#Carregar a base de dados a fim de que possam ser transformados diretamente para tensores
transform = transforms.ToTensor()
```

```python
#Leitura dos dados
train = datasets.MNIST(root = '.', download = True, train = True, transform = transform)
test = datasets.MNIST(root = '.', download = True, train = False, transform = transform)
```

```python
#Printamos o tamanho da nossa base de dados de treino e teste, assim como os respectivos tamanhos de pixels de largura e altura
print(train.data.shape, test.data.shape)
```

```python
#Visualizaremos nosso primeiro registro
print(train.data[0])
```

```python
'''
Para que a base de dados seja normalizada, utilizamos o método loader, 
no qual será responsável por normalizar nossos dados entre zero e um
'''
train_loader = torch.utils.data.DataLoader(train, batch_size=128)
test_loader = torch.utils.data.DataLoader(test, batch_size=128)
```

```python
class classificador(nn.Module):
    def __init__ (self):
        super().__init__()
        '''Logo abaixo iremos determinar nossa primeira camada convolucinal, no qual definimos nossa convolução
        2d e logo após determinamos o tipo de imagem, seja RGB ou não (no nosso caso não), assim como aplicamos o número de
        filtros (32), ou seja, trinta e dois canais de saída. 
        '''
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=(3,3))
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=32, kernel_size=(3,3))
        self.activation = nn.ReLU() #Nossa função de ativação
        self.bnorm = nn.BatchNorm2d(num_features=32) #Aplicamos a normalização dos dados dentro das camadas ocultas
        self.pool = nn.MaxPool2d(kernel_size=(2,2)) #Kernel size será nossa matriz 2x2 capaz de identificar padrões e reduzir dimensões
        self.flatten = nn.Flatten()
        '''
            output = (input - filter + 1)/stride
        Convolução 1: (28 - 3 + 1)/1 = 26x26
        pooling 1: 26/2 = 13x13
        
        Convolução 2: (13 - 3 + 1)/ 1 = 11x11
        pooling 2: 11/2 = 5x5
        '''
        self.linear1 = nn.Linear(in_features=32*5*5, out_features=128)
        self.linear2 = nn.Linear(in_features=128, out_features=128)
        self.output = nn.Linear(in_features=128, out_features=10)
        self.dropout = nn.Dropout(p = 0.2)
        
    def forward(self, X):
        X = self.pool(self.bnorm(self.activation(self.conv1(X))))
        X = self.pool(self.bnorm(self.activation(self.conv2(X))))
        X = self.flatten(X)
        
        X = self.dropout(self.activation(self.linear1(X)))
        X = self.dropout(self.activation(self.linear2(X)))
        X = self.output(X)
        
        return X
```

```python
net = classificador()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(net.parameters())
```

```python
#Setamos a GPU necessária para realizar nossa classificação
device = torch.device("cuda") if torch.cuda.is_available else torch.device("cpu")
print(device)

net.to(device)
```

```python
def training_loop(loader, epoch):
    running_loss = 0.
    running_accuracy = 0.
    for i, data in enumerate(loader):
        
        inputs, labels = data
        
        inputs, labels = inputs.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = net(inputs)
        
        loss = criterion(outputs, labels)
        loss.backward()
        
        optimizer.step()
        
        running_loss += loss.item()
        
        ps = F.softmax(outputs, dim=1)
        top_p, top_class = ps.topk(k=1, dim=1)
        equals = top_class == labels.view(*top_class.shape)
        
        accuracy= torch.mean(equals.type(torch.float))
        
        running_accuracy += accuracy
        
        #Imprimir os dados referentes a este loop
        print("Época {:3d} - Loop {:3d} de {:3d}: perda {:03.2f} - precisão {:03.2f}".format(epoch + 1, i+1, len(loader)\
                                                                                             , loss, accuracy), end= '\r')
    #Impressão referente à épocas
    print("Época {:3d} Finalizada: perda {:.5f} - precisão {:.5f}".format(epoch+1, running_loss/len(loader)\
                                                                         ,running_accuracy/len(loader)))
```

```python
for epoch in range(5):
    print("Treinando...")
    training_loop(train_loader, epoch)
    net.eval()
    print('Validando...')
    training_loop(test_loader, epoch)
    net.train()
```
